# **Segmentación de instancias de vídeos con PixelLib**

Para realizar la segmentación de instancias con PixelLib:
1. Descargar el modelo de máscara rcnn desde [aquí](https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.2/mask_rcnn_coco.h5).
2. Crear carpeta en drive y cargar el modelo.
3. Cargar video de prueba en carpeta de drive.
4. Montar drive en almacenamiento de sesión.

Importamos la clase `instance_segmentation` para realizar la segmentación de instancias y creamos una instancia de la clase.
Cargamos el modelo maskrcnn entrenado en el conjunto de datos de COCO.
Llamamos a la función para realizar la segmentación en el archivo de vídeo.

Toma los siguientes parámetros:-

`video_path`: la ruta al archivo de vídeo en la que queremos realizar la segmentación.

`frames_per_second`: este es un parámetro para establecer el número de fotogramas por segundo para el archivo de vídeo de salida. En este caso se establece en 15, es decir, el archivo de vídeo guardado tendrá 15 fotogramas por segundo.

`output_video_name`: el vídeo segmentado guardado. El vídeo de salida se guardará en el directorio de trabajo actual.

```
import pixellib
from pixellib.instance import instance_segmentation

segment_video = instance_segmentation()
segment_video.load_model("/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/mask_rcnn_coco.h5")
segment_video.process_video("/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/video_prueba.mp4", frames_per_second= 20, output_video_name="/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/output_video.mp4")
```

Podemos realizar la segmentación de instancias con la detección de objetos estableciendo el parámetro `show_bboxes` en `true`.

```
import pixellib
from pixellib.instance import instance_segmentation

segment_video = instance_segmentation()
segment_video.load_model("/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/mask_rcnn_coco.h5")
segment_video.process_video("/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/video_prueba.mp4", show_bboxes = True, frames_per_second= 20, output_video_name="/content/drive/My Drive/2020-35 Pixellib/Modelo_preentrenado/output_video_2.mp4")
```
